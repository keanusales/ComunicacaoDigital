# Atividade 4 - CRC

- String a se transmitir D: 1101011011
- Gerador G(x): 10011

#### Importando da extensão em C

```python
from crc import transfer, receiver
```

#### Execução do código

```python
data, gen = 0b1101011011, 0b10011
res1 = transfer(data, gen)
print(f"Resposta: {bin(res1)}")
res2 = receiver(res1, gen)
print(f"Checagem: {bin(res2)}")
```
