# Atividade 4 - CRC

- String a se transmitir D: 1101011011
- Gerador G(x): 10011

#### Importando da extensão em C

```python
from crc import transfer, receiver
```

#### Execução do código

```python
pregen = 0b10011
data = 0b1101011011
res = transfer(data, pregen)
print(f"Resposta: {bin(res)}")
res = receiver(res, pregen)
print(f"Checagem: {bin(res)}")
```
