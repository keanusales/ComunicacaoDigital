# Atividade 4 - CRC

- String a se transmitir D: 1101011011
- Gerador G(x): 10011

#### Cálculo do transmissor

```python
def transfer(data: int, gen: int):
  odata = data.bit_length() - 1
  orgen = gen.bit_length() - 1
  smord = odata + orgen
  send = rest = data << orgen
  for i in range(odata + 1):
    if rest & (1 << (smord - i)):
      rest ^= gen << (odata - i)
  return send ^ rest
```

#### Cálculo do receptor

```python
def receiver(data: int, gen: int):
  odata = data.bit_length() - 1
  orgen = gen.bit_length() - 1
  smord = odata - orgen
  for i in range(smord + 1):
    if data & (1 << (odata - i)):
      data ^= gen << (smord - i)
  return data
```

#### Execução do código

```python
data, gen = 0b1101011011, 0b10011
res1 = transfer(data, gen)
print(f"Resposta: {bin(res1)}")
res2 = receiver(res1, gen)
print(f"Checagem: {bin(res2)}")
```
